## 1. Ответы в листьях регрессионного дерева.

Рассматриваем произвольный лист. Пусть для объектов выборки $x_1^{(0)}, \dots x_n^{(0)}$, попавших в данный лист, таргеты равны $y_1^{(0)}, \dots y_n^{(0)}$.

В первом случае мы отвечаем средним значением: $M = \overline{y^{(0)}}$
Во втором - случайной величной $\xi_i$ б принимающей равномерно значения $y_1^{(0)}, \dots y_n^{(0)}$.

Запишем матожидание MSE, где $y$ - истинное значение таргетов, $\hat{y}$ - используемая оценка :
$E ( L(y,\hat{y}) ) = E (\sum (y_i - \hat{y_i})^2) = \sum E(y_i - \hat{y_i})^2 = \sum E y_i^2 - 2\sum E y_i E \hat{y_i} + \sum E \hat {y_i} ^ 2$

Для обоих случаев 1-ые слагаемые совпадают, так как не зависит от оценки. Для 2-го:

1): $\forall i: E \hat{y_i} = E M = M$

2): $\forall i: E \hat{y_i} = E \xi_i =  \overline{y^{(0)}} = M$

Поэтому 2-ые слагаемые также совпадают.
Рассмотрим 3-е:

1):  $\forall i: E \hat{y_i}^2 = E M^2 = M^2$

2):  $\forall i: E \hat{y_i}^2 = E \xi^2 \geq (E \xi)^2 = M^2$

То есть, в среднем вторая оценка не лучше первой, значит лучше отвечать средним значением на листе

## 2. Линейные модели в деревьях.

Если реальная модель не линейная, то линейная регрессия в листах не должна давать ощутимого преимущества.

Возможна ситуация, что данные разбиваются на классы в которых на подмножестве признаков модель линейная. Если найти такие признаки, и построить дерево по оставшимся признакам (чтобы попытаться разбить на такие классы), а в листьях применить регрессию на этих признаках, то это может дать выигрыш.
