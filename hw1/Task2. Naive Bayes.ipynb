```python
from sklearn import cross_validation, datasets, metrics, neighbors 
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
digits = datasets.load_digits()
breast_cancer = datasets.load_breast_cancer()
```

```python
print(digits['data'][:5])
print(breast_cancer['data'][:5])
```

```python
def quality(X, y):
    clfs = [GaussianNB(), MultinomialNB(), BernoulliNB()]
    methods = ['GaussianNB', 'MultinomialNB', 'BernoulliNB']
    qa = []
    for clf in clfs:
        qa.append(cross_validation.cross_val_score(clf, X, y).mean())
    for i in range(len(qa)):
        print(methods[i], ': ', qa[i])
```

```python
print('digits:')
quality(digits['data'], digits['target'])
print('breast cancer:')
quality(breast_cancer['data'], breast_cancer['target'])
```

В случае с digits признаки дискретны и вполне понятно, почему multinomialNB отработал лучше. (Качество на кросс-валидации 0,87)

В датасете breast_canser признаки распределены нормально (ну более-менее), в таких случаях gaussianNB эффективнее. (Качество на кросс-валидации 0,936)

Верны утверждения (c) и (d): На целых неотрицательных признаках лучше отработало мультиномиальное распределение, на вещественных - нормальное.
